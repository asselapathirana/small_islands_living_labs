```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point
```

```python
  # Load the Excel file again
population_data = pd.read_excel('nucleos-de-poblacion.xlsx')

# Creating points for GeoDataFrame
points = [Point(xy) for xy in zip(population_data['NP_XLON'], population_data['NP_YLAT'])]
geo_df = gpd.GeoDataFrame(population_data, geometry=points)

# Load map of Spain
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
spain = world[world.name == "Spain"]

# Plot
fig, ax = plt.subplots(figsize=(10, 15))
spain.plot(ax=ax, color='white', edgecolor='black')
geo_df.plot(ax=ax, markersize=geo_df['TOTAL']/100, color='blue', alpha=0.5)

plt.title('Population Centers in Spain')
plt.ylabel('Latitude')
plt.show()
```
